```python
# Load GTFS data into DataFrames
feed = Feed(r'C:\Users\sd\OneDrive\Indicateur TC\GTFS.zip',time_windows=[0, 6, 9, 15, 19, 22, 24], busiest_date= False)
calendar_df = feed.calendar_dates
stop_times_df = feed.stop_times
```

```python
from datetime import timedelta
```

```python
# Merge the data to have the dates on it
merged_data = pd.merge(stop_times_df, calendar_df, on='service_id')

# Convert all dates to a consistent format (e.g., YYYY-MM-DD)
merged_data['date'] = pd.to_datetime(merged_data['date'], format='%Y%m%d', errors='coerce')
```

```python
# Step 1: Convert 'arrival_time' column to timedeltas
merged_data['arrival_time'] = merged_data['arrival_time'].apply(lambda x: timedelta(seconds=x))

# Step 2: Subtract the timedelta for 3:45 AM
start_time = timedelta(hours=3, minutes=45)
merged_data['adjusted_time'] = merged_data['arrival_time'] - start_time

# Step 3: Combine hours, minutes, and seconds into a single column
merged_data['time_in_hhmmss'] = merged_data['adjusted_time'].apply(lambda x: str(x))

merged_data[['time_in_hhmmss']]

```

```python
table_dates = merged_data[['parent_station','date','arrival_time','stop_id','stop_name']].groupby(['parent_station','date']).agg(
    arrival_time_max=pd.NamedAgg(column="arrival_time", aggfunc="max"),
    arrival_time_min=pd.NamedAgg(column="arrival_time", aggfunc="min"),
    stop_name=pd.NamedAgg(column="stop_name", aggfunc="max")).reset_index()
```

```python
table_dates['Amplitude en minutes'] = ((table_dates['arrival_time_max'] - table_dates['arrival_time_min']) / (60*10080)*100)
table_dates
```

```python
# Parameteres su ArcGIS tool
date_debut = '2023-08-07'
date_fin = '2023-08-13'

# Flter to have only trips from date_debut to date_fin
Trips_per_week7 = table_dates[(table_dates['date']>= date_debut) & (table_dates['date']<= date_fin)]
Trips_per_week7
```

```python
Result = Trips_per_week7[['parent_station','stop_name','Amplitude en minutes']].groupby(['parent_station','stop_name']).sum().reset_index()
Result
```
