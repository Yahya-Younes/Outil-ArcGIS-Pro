```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('RueAlsaceLorraine.csv')

# Extract the IDs and their corresponding notes
ids = df['LINK_ID'].unique()
indicateurs = ['CIR_MOY','Note_NbVoie', 'Note_ArretTP', 'Note_Obstacle', 'Note_Carrefour', 'Note_Vitesse', 'Note_Congestion', 'Note_Chantier','ACC_MOY', 'Note_Gabarit', 'Note_Stationnement', 'Note_Horaire', 'Note_Pente']

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Define colors and marker styles for each ID
id_styles = {
    ids[0]: ('green', 'o'),
    ids[1]: ('orange', '^'),
    ids[2]: ('red', 's')
}

# Reverse the order of ids for legend convenience
ids = ids[::-1]

# Plot the notes for each ID with different colors and marker styles
for idx, id_ in enumerate(ids):
    id_data = df[df['LINK_ID'] == id_]
    notes = id_data[indicateurs].values.flatten()
    color, marker = id_styles[id_]
    ax.plot(indicateurs, notes, marker=marker, color=color, markersize=8, linestyle='-', label=f'LINK_ID {id_}')

# Set the y-axis ticks and labels for the three possibilities (1, 2, 3)
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(['1', '2', '3'])

# Set the plot title and labels
ax.set_title('Notes de 3 tronçons de la rue Alsace-Lorraine')
ax.set_xlabel('Indicateurs')
ax.set_ylabel('Note attribuée')

# Add a legend to distinguish the different IDs
ax.legend()

# Customize the x-axis labels to make only two labels bold and bigger
for label in ax.get_xticklabels():
    if label.get_text() in ['CIR_MOY', 'ACC_MOY']:
        label.set_fontsize(26)
        label.set_fontweight('bold')

# Show the plot
plt.tight_layout()
plt.grid(visible=True)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('TableRuesSonia.csv')

# Group the data by 'ST_NAME'
df_grouped = data.groupby('ST_NAME')

# Extract the unique indicators
indicateurs = ['CIR_MOY', 'Note_NbVoie', 'Note_ArretTP', 'Note_Obstacle', 'Note_Carrefour', 'Note_Vitesse', 'Note_Congestion', 'Note_Chantier', 'ACC_MOY', 'Note_Gabarit', 'Note_Stationnement', 'Note_Horaire', 'Note_Pente']

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Define a custom colormap from red to green
custom_cmap = mcolors.LinearSegmentedColormap.from_list('custom_color_map', [(0, 'red'), (0.5, 'yellow'), (1, 'green')])

# Plot the data for each ST_NAME with colors ranging from red to green based on note values
for idx, (st_name, group_df) in enumerate(df_grouped):
    notes = group_df[indicateurs].mean()  # Calculate the mean for each indicator
    normalized_notes = (notes - 1) / 2  # Normalize note values to range between 0 and 1
    colors = custom_cmap(normalized_notes)  # Get the colors based on normalized note values

    # Plot each indicator with the respective colors
    for i in range(len(indicateurs)):
        ax.scatter(indicateurs[i], notes[i], marker='o', color=colors[i], s=100, label=st_name)

# Set the y-axis ticks and labels for the three possibilities (1, 2, 3)
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(['1', '2', '3'])

# Set the plot title and labels
ax.set_title('Notes en détail de chaque rue')
ax.set_xlabel('Indicateurs')
ax.set_ylabel('Note attribuée')

# Add a legend to distinguish the different ST_NAMEs
ax.legend()

# Customize the x-axis labels to make only two labels bold and bigger
for label in ax.get_xticklabels():
    if label.get_text() in ['CIR_MOY', 'ACC_MOY']:
        label.set_fontsize(14)
        label.set_fontweight('bold')

# Show the plot
plt.tight_layout()
plt.xticks(rotation=45)
plt.grid(visible=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('TableRuesSonia.csv')

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Define colors and marker styles for each street (ST_NAME)
num_streets = len(data['ST_NAME'].unique())
colors = plt.cm.tab20.colors[:num_streets]
markers = ['o', '^', 's', 'd', 'x', 'P', 'h', '+', 'v', '<', '>', '*']

# Plot the data for each street with different colors and marker styles
for idx, (st_name, group_df) in enumerate(data.groupby('ST_NAME')):
    color = colors[idx]
    marker = markers[idx % len(markers)]
    
    # Plot the data for each LINK_ID within the current ST_NAME
    for link_id, link_data in group_df.groupby('LINK_ID'):
        notes = link_data[indicateurs]  # Extract the notes for the current LINK_ID
        ax.plot(indicateurs, notes, marker=marker, color=color, markersize=8, linestyle='-', label=f'{st_name} - LINK_ID: {link_id}')

# Set the y-axis ticks and labels for the three possibilities (1, 2, 3)
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(['1', '2', '3'])

# Set the plot title and labels
ax.set_title('Notes en détail de chaque rue')
ax.set_xlabel('Indicateurs')
ax.set_ylabel('Note attribuée')

# Add a legend to distinguish the different ST_NAMEs and LINK_IDs
ax.legend()

# Customize the x-axis labels to make only two labels bold and bigger
for label in ax.get_xticklabels():
    if label.get_text() in ['CIR_MOY', 'ACC_MOY']:
        label.set_fontsize(14)
        label.set_fontweight('bold')

# Show the plot
plt.xticks(rotation=45)
plt.grid(visible=True)
plt.tight_layout()
plt.show()
```
